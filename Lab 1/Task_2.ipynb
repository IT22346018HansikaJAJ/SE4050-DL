```python
import pandas as pd

# Load the dataset
df = pd.read_csv("/content/drive/MyDrive/Jana/SE4050-DL-Lecture02-main/SE4050-DL-Lecture02-main/Au_nanoparticle_dataset.csv")

# Show the first five rows
df.head()
```

```python
new_df.head(20)
```

# 3. Calculate mean, std, and quartiles

```python
# Mean
mean_values = new_df.mean()
print("Mean:\n", mean_values)

# Standard Deviation
std_values = new_df.std()
print("\nStandard Deviation:\n", std_values)

# Quartiles
quartiles = new_df.quantile([0.25, 0.5, 0.75])
print("\nQuartiles:\n", quartiles)
```

# 4. Plot histograms in a 1x4 layout

```python
import matplotlib.pyplot as plt

features = ['N_total', 'N_bulk', 'N_surface', 'R_avg']
plt.figure(figsize=(20, 4))

for i, feature in enumerate(features):
    plt.subplot(1, 4, i+1)
    plt.hist(new_df[feature], bins=30, color='skyblue', edgecolor='black')
    plt.title(f'Histogram of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

# 5. Visualize scatter plots and histograms using sns.pairplot()

```python
import seaborn as sns

sns.pairplot(new_df, diag_kind="hist")
plt.suptitle("Pairplot of Selected Features", y=1.02)
plt.show()
```

# 6. Customized PairGrid with diagonal hist+KDE, lower KDE, upper histograms

```python
g = sns.PairGrid(new_df)

# Upper triangle: 2D histogram
g.map_upper(sns.histplot)

# Diagonal: histogram + KDE
g.map_diag(sns.histplot, kde=True)

# Lower triangle: bivariate KDE
g.map_lower(sns.kdeplot, fill=True, cmap="Blues")

plt.suptitle("Customized PairGrid with KDE and Histograms", y=1.02)
plt.show()
```
