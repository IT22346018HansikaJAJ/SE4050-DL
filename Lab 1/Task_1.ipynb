# 1. Create a random 4x4 array from an exponential distribution

```python
import numpy as np

# Exponential distribution with scale (lambda=1.0)
exp_array = np.random.exponential(scale=1.0, size=(4, 4))
print("4x4 Exponential Array:\n", exp_array)
```

# 2. Create a 100000x1 exponential array and plot histogram along with uniform and normal distributions

```python
import matplotlib.pyplot as plt

# Create 100000x1 arrays
exp_data = np.random.exponential(scale=1.0, size=100000)
uniform_data = np.random.rand(100000)
normal_data = np.random.randn(100000)

# Plot histograms
plt.figure(figsize=(10, 6))
plt.hist(uniform_data, density=True, bins=100, histtype="step", color="blue", label="Uniform")
plt.hist(normal_data, density=True, bins=100, histtype="step", color="green", label="Normal")
plt.hist(exp_data, density=True, bins=100, histtype="step", color="red", label="Exponential")

plt.axis([-2.5, 8, 0, 1.1])
plt.legend(loc="upper right")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```

# 4. Pearson and Spearman Correlation for Pok√©mon Stats with Heatmaps

```python
import pandas as pd

!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
# Read dataset
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Drop rows with missing values
df = df.dropna()

# Selecting relevant features
features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
data = df[features]

# Pearson correlation
pearson_corr = data.corr(method='pearson')

# Spearman correlation
spearman_corr = data.corr(method='spearman')

# Heatmap for Pearson
plt.figure(figsize=(8, 6))
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Pearson Correlation Heatmap")
plt.show()

# Heatmap for Spearman
plt.figure(figsize=(8, 6))
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Spearman Correlation Heatmap")
plt.show()
```
